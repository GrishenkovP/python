```python
import pandas as pd
import itertools
import random
import hashlib
from tqdm import tqdm_notebook
```

```python
list_check_number = []
list_product_name = []
for _ in tqdm_notebook(range(1000000)):
  list_check_number.append(random.randint(1,350000))

  current_product = "pr_" + str(random.randint(1,5000))
  list_product_name.append(current_product)
```

```python
d_demo_data = {'check_number': list_check_number, 'product_name': list_product_name}
```

```python
df_sales = pd.DataFrame(data=d_demo_data)
```

```python
df_sales = df_sales.sort_values(by=['check_number','product_name'])
df_sales.head()
```

```python
df_sales_group = df_sales.groupby(by=['check_number'])['product_name'].apply(list).reset_index(name='list_product_name')
```

```python
df_sales_group.head(25)
```

```python
d_code_couple_products = {} 
d_amount_couple_products = {}
```

```python
for current_list in tqdm_notebook(df_sales_group['list_product_name']):
  current_list_drop_duplicates = list(set(current_list))
  list_combinations = list(itertools.combinations(current_list_drop_duplicates, 2))
  for element_list_combinations in list_combinations:
    if element_list_combinations[0] == element_list_combinations[1]:
      continue
    else:
      list_sorted = sorted(element_list_combinations)
      str_sorted = ", ".join(list_sorted)
      hash_object = hashlib.md5(str_sorted.encode('utf-8'))
      hash_code = hash_object.hexdigest()
      if d_code_couple_products.get(hash_code) == None:
        d_code_couple_products[hash_code] = list_sorted
        d_amount_couple_products[hash_code] = 1
      else:
        d_amount_couple_products[hash_code] = d_amount_couple_products[hash_code] + 1
```

```python
list_couple_products = []
list_amount = []
```

```python
for current_key in tqdm_notebook(d_code_couple_products.keys()):
  list_couple_products.append(d_code_couple_products.get(current_key))
  list_amount.append(d_amount_couple_products.get(current_key))
```

```python
d_data = {'two_products':list_couple_products, 'amount':list_amount}
```

```python
df_final = pd.DataFrame(data=d_data)
df_final = df_final.sort_values(by=['amount'], ascending=False)
```
