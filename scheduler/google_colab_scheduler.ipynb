```python
import schedule 
import time
```

```python
def test_task(name):
    print('Hello', name)
```

```python
schedule.every(5).seconds.do(test_task, name='World!').tag('test_job')
```

```python
while True:
    schedule.run_pending()
    time.sleep(1)
```

```python
#Список jobs
schedule.jobs
```

```python
#Удаление job
schedule.clear('test_job')
schedule.jobs
```
